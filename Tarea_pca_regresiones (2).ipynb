# **NORMALIZACION DEL CONJUNTO DE DATOS**
Es obligatorio estandarizar el conjunto de datos antes de aplicar PCA, de lo contrario, producir resultados errneos. 

Aqu utilizamos la funcin StandardScaler() del mdulo sklearn.preprocessing para estandarizar los conjuntos de datos 

de entrenamiento y de prueba.

```python
# this is an optional step
df_st =  StandardScaler().fit_transform(data0)  
pd.DataFrame(df_st, columns=data0.columns).head()
```

# **ESTANDARIZACION**

El conjunto de datos de estandarizacin con escala (media=0, varianza=1) es necesario ya que elimina los sesgos de las variables originales. Por ejemplo, cuando los datos de cada variable se recogen en unidades diferentes.
Las variables estandarizadas no tendrn unidades y tendrn una varianza similar.
La estandarizacin es un mtodo aconsejable para la transformacin de datos cuando las variables del conjunto de datos original se han medido en una escala significativamente diferente.
En algunos casos, no es necesario estandarizar el conjunto de datos, ya que la variacin original del conjunto de datos es importante (Gewers et al., 2018)

# **REALIZAR PCA MEDIANTE SCIKIT-LEARN**

```python
pca_out = PCA(n_components=16).fit(df_st)

# get the component variance
# Proportion of Variance (from PC1 to PC16)
pca_out.explained_variance_ratio_
```

```python
# Cumulative proportion of variance (from PC1 to PC16)   
np.cumsum(pca_out.explained_variance_ratio_)
```

# **COEFICIENTE DE CORRELACIN ENTRE LAS VARIABLES ORIGINALES Y EL COMPONENTE**

```python
# component loadings or weights (correlation coefficient between original variables and the component) 
# component loadings represents the elements of the eigenvector
# the squared loadings within the PCs always sums to 1
loadings = pca_out.components_
num_pc = pca_out.n_features_
pc_list = ["PC"+str(i) for i in list(range(1, num_pc+1))]
loadings_df = pd.DataFrame.from_dict(dict(zip(pc_list, loadings)))
loadings_df['variable'] = data0.columns.values
loadings_df = loadings_df.set_index('variable')
loadings_df
```

# **RETENCIN DE LOS COMPONENTES PRINCIPALES (PC)**


Cuando el nmero de PCs es igual al nmero de variables originales, deberamos mantener slo los PCs que explican la mayor parte de la varianza (70-95%) para facilitar la interpretacin. Cuantas ms PCs se incluyan que expliquen la mayor parte de la variacin en los datos originales, mejor ser el modelo PCA. Esto es muy subjetivo y se basa en la interpretacin del usuario (Cangelosi et al., 2007).

Los valores propios (varianza explicada por cada PC) de las PC pueden ayudar a retener el nmero de PC. Generalmente, los PC con valores propios > 1 contribuyen a una mayor varianza y deberan retenerse para el anlisis posterior.

El grfico Scree (para la prueba del codo) es otra tcnica grfica til para retener las PC. Deberamos mantener los PCs en los que hay un cambio brusco en la pendiente de la lnea que conecta los PCs adyacentes.

# **PROPORSIN DE VARIANZAS**

```python
plt.figure(figsize=(12,5))
plt.bar(loadings_df.columns,pca_out.explained_variance_ratio_*100,color='green', width=0.8)
plt.ylabel("Proportion of variance (%)")
plt.xlabel("PCs")
plt.show()
```

# **PROPORSIN ACUMULADA DE VARIANZAS**

```python
plt.figure(figsize=(12,5),)
plt.bar(loadings_df.columns,np.cumsum(pca_out.explained_variance_ratio_*100),color='red', width=0.6)
plt.ylabel("Cumulative proportion of variance (%)")
plt.xlabel("PCs")
plt.show()
```

```python
# get eigenvalues (variance explained by each PC)  
pca_out.explained_variance_

# get scree plot (for scree or elbow test)
from bioinfokit.visuz import cluster
cluster.screeplot(obj=[loadings_df.columns,pca_out.explained_variance_ratio_],show=True,dim=(12,5),axlabelfontsize=16)
plt.show()
# Scree plot will be saved in the same directory with name screeplot.png
```

```python
# get PCA loadings plots (2D and 3D)
# 2D
cluster.pcaplot(x=loadings[0], y=loadings[1], labels=data0.columns.values,
                show=True,dim=(20,15),axlabelfontsize=14,
    var1=round(pca_out.explained_variance_ratio_[0]*100, 2),
    var2=round(pca_out.explained_variance_ratio_[1]*100, 2))
```
